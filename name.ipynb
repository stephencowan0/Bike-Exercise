```python
import pandas as pd
import numpy as np

dfData = pd.read_csv("Data.csv")
dfData['Date'] = pd.to_datetime(dfData['Date'], format='%Y%m%d')
dfData['Date'] = dfData['Date'].dt.strftime('%m-%d')
dfData
```

```python
import matplotlib.pylab as plt
%matplotlib inline

plt.style.use('dark_background')
plt.plot(dfData['Date'], dfData['Duration'], color="blue")
plt.title('Duration')
plt.xlabel('Date')
plt.ylabel('Minutes')
plt.grid(True)
plt.show()

plt.plot(dfData['Date'], dfData['Distance(MI)'], color="blue")
plt.title('Distance')
plt.xlabel('Distance')
plt.ylabel('Miles')
plt.grid(True)
plt.show()

plt.plot(dfData['Date'], dfData['KCAL'], color="blue")
plt.title('KCAL')
plt.xlabel('Date')
plt.ylabel('KCAL')
plt.grid(True)
plt.show()

plt.plot(dfData['Date'], dfData['KCAL/HR'], color="blue")
plt.title('KCAL/HR')
plt.xlabel('Date')
plt.ylabel('KCAL/HR')
plt.grid(True)
plt.show()

plt.plot(dfData['Date'], dfData['MPH'], color="blue")
plt.title('Miles Per Hour')
plt.xlabel('Date')
plt.ylabel('MPH')
plt.grid(True)
plt.show()

plt.plot(dfData['Date'], dfData['RPM'], color="blue")
plt.title('Rotations Per Minutes')
plt.xlabel('Date')
plt.ylabel('RPM')
plt.grid(True)
plt.show()

plt.plot(dfData['Date'], dfData['Watts'], color="blue")
plt.title('Watts')
plt.xlabel('Date')
plt.ylabel('Watts')
plt.grid(True)
plt.show()
```
